### Reddit 

* https://lifehacker.com/a-beginners-guide-to-reddit-1798643829
    * Based on the idea to let the online community decide what is newsworthy and what it's not (very best would rise to the top)
    * 1 billion users per month 
    * site can seem a little overwhelming 
    * reddiquette = particular set of guidelines by which they ask participants to behave 
        * Dislike of self-promotion 
        * Ban on submitting a lot of links very quickly 
        * Looking for the original source of a story rather than linking to a blog that has picked it up from another blog 
        * Looking for duplicates of your story 
        * Posting to the most appropriate subreddit possible 
        * Acting like a decent person -> remember the human (maxium)
    * Subreddit = individual message board devoted to one particular topic (there are thousands of them)
        * All subreddit names begin with a lowercase "r" followed by a slash (e.g., r/news
        * Each subreddit is hosted by a moderator and comes with its own rules reagarding posts 
            * Any link you share in r/science must be less than six months old
        * You can subscribe (join) subreddits
        * Posts on subreddits are arranged according to how many upvotes they have received (but Reddit will also give recent posts a prominent place in the feed in order to give them a chance to be seen and voted. 
        * Gray number next to each post represents the post's rank (number of upvotes - number of downvotes)
            * Same system applies to comments as well. 
    * Homepage lists posts from the subreddits you follow
    * You can create three types of posts 
        1. Share a link
        2. Post a video
        3. Post an image 
        4. Text post (original thought of your own)
